```python
def open_day(day=1):
    return [int(x.strip()) for x in open(f"../data/day{day}.txt", "r").readlines()]
data = open_day(1)
# print(data)
```

```python
def part_one(data, verbose=False):
    counter = 0
    for index in range(len(data)):
        if not index: continue
        first_thing = data[index-1]
        second_thing = data[index]
        if second_thing > first_thing:
            counter += 1
        # report!
        if verbose: print(f"{index}| {first_thing} {'>' if first_thing>second_thing else '<'} {second_thing}")
        # break
    if verbose: print(f"There were {len(data)} elements. There are {counter} measurements that are larger than the previous measurement.")
    return counter
part_one(data)
```

```python
def part_two(data, verbose=True):
    # rebuild data to be a 3 element sliding window.
    new_data = [sum([data[i], data[i+1], data[i+2]]) for i in range(len(data)-2)]
    print(new_data)
    return part_one(new_data, True)
part_two(data, True)
```
